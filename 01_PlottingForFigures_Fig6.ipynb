# Degree Distributions

## Fig 6a

```python
SysColors={"gas":'black', "fluid":'blue', "solid":'red'}
PlotLabels={"gas":'Vapor', "fluid":'Fluid', "solid":'Solid'}
```

```python
myFig, myAx = plt.subplots(1,1, constrained_layout=True, figsize=[4.0,4.0/1.2])

for tractID, aTract in enumerate(['gas', 'fluid', 'solid']):
    tmp_data = np.loadtxt(f"02_degrees/data/Degs_lj_{aTract}.txt")
    plt.step(tmp_data[0], tmp_data[1], where='mid', lw=2, label=PlotLabels[aTract], 
             color=SysColors[aTract], alpha=1.0)
    plt.errorbar(*tmp_data, fmt='none', color=SysColors[aTract], capsize=2, capthick=1.5)
    
plt.xlim(0,20)
plt.yscale('log')
plt.ylim(1e-4, 1.2e0)
myFig.legend(bbox_to_anchor=(1.0, 0.55), loc='center left', title='LJ Phase')
plt.xlabel("$k$ (node degree)")
plt.ylabel("$P(k)$ (degree distribution)")
plt.savefig('figures/Fig7a.pdf', transparent=True, bbox_inches='tight')
plt.savefig('figures/Fig7a.png', transparent=True, bbox_inches='tight')
plt.show()
```

## Fig 6b

```python
TractColors={'A0':'b','A1':'g','A2':'m'}
TractLabels={f'A{i}':f'A$_{{{i}}}$-rpL5' for i in range(3)}
```

```python
myFig, myAx = plt.subplots(1,1, constrained_layout=True, figsize=[4,4/1.2])

for tractID, aTract in enumerate(['A0', 'A1', 'A2']):
    tmp_data = np.loadtxt(f"02_degrees/data/Degs_lk_8_{aTract}.txt")
    plt.step(tmp_data[0], tmp_data[1], where='mid', lw=2, label=TractLabels[aTract], 
             color=TractColors[aTract], alpha=1.0)
    plt.errorbar(*tmp_data, fmt='none', color=TractColors[aTract], capsize=2, capthick=1.5)
    
plt.xlim(0,40)
plt.yscale('log')
plt.ylim(1e-4, 1e0)
myFig.legend(bbox_to_anchor=(0.975, 0.975), loc='upper right', title='Network')
plt.xlabel("$k$ (node degree)")
plt.ylabel("$P(k)$ (degree distribution)")
plt.savefig('figures/Fig7b.pdf', transparent=True, bbox_inches='tight')
plt.savefig('figures/Fig7b.png', transparent=True, bbox_inches='tight')
plt.show()
```
